```python
# импорт библиотек
import pandas as pd
import numpy as np
import seaborn as sns
```

## чтение файлы данных 

```python
df = pd.read_csv("https://eddydewrussia.ru/download/muzik_q2_21/?wpdmdl=5083&refresh=62869d1b40fc81652989211", encoding = 'cp1251', sep=';')
df.head()
```

```python
# получаю информацию о таблице
df.info()
```

### первое впечатление
* странное количество строк в первой колонке
* есть колонка с пустыми значениями
* названия колонок не в змеином_регистре

```python
df.head()
```

```python
df.tail()
```

### видно, что последняя строка лишняя. её лучше удалить

```python
df.drop(df.tail(1).index,inplace=True)
df.tail()
```

### удаляю пустую колонку

```python
del df['Art.No.'] 
df.head()
```

### перевожу названия колонок в змеиный_регистр

```python
df = df.rename(columns={'Labelname': 'labelname', 'ISRC': 'isrc', 'EAN/UPC': 'ean_upc', 
                        'Artist': 'artist', 'Producttitle': 'producttitle', 'Tracktitle': 'tracktitle',
                        'Outletname': 'outletname', 'Format': 'format', 'Territory': 'territory',
                        'Sales Period': 'sales_period', 'Units': 'units',
                        'Royalty Amount Customer': 'royalty_amount_customer'})
df.head()
```

### проверка на уникальность

```python
for col in list(df):
    print(col)
    print(np.sort(df[col].unique()))
```

### неявные дубликаты
* producttitle 'Ой, мороз, мороз.' и 'Ой; мороз; мороз.'
* tracktitle 'Ой, мороз, мороз.' и 'Ой; мороз; мороз.'

### замена неявных дубликатов

```python
df['producttitle'] = df['producttitle'].replace('Ой; мороз; мороз.', 'Ой, мороз, мороз.')
df['tracktitle'] = df['tracktitle'].replace('Ой; мороз; мороз.', 'Ой, мороз, мороз.')
```

```python
# снова проверка неявных дублиткатов
for col in list(df):
    print(col)
    print(np.sort(df[col].unique()))
```

### дубликатов больше нет


### проверяю формат числовых данных
* это колонки Units и royalty_amount_customer

```python
df.info()
print('units', type(df['units'][5]))
print('royalty_amount_customer', type(df['royalty_amount_customer'][5]))
```

### тип данных royalty_amount_customer строка
* перевожу его в числовой

```python
# замена "," на "."
df['royalty_amount_customer'] = df['royalty_amount_customer'].str.replace(',','.')
df.head()

```

```python
print('royalty_amount_customer', type(df['royalty_amount_customer'][5]))
```

```python
# меняю тип данных со строчных на численные
df['royalty_amount_customer'] = df['royalty_amount_customer'].astype('float')

# проверяю тип данных
print('royalty_amount_customer', type(df['royalty_amount_customer'][5]))
```

```python
df.head()
```

### данные подготовлены для анализа

## Выявить самые привлекательные по цене
* суммы units по площадкам

```python
units_sum_by_outletname = df.groupby('outletname')['units'].sum()
print(units_sum_by_outletname)
```

```python
# суммы units по royalti в увеличенном в 1000000 раз для удобства восприятия
units_sum_by_royaly = df.groupby('outletname')['royalty_amount_customer'].sum() * 100 * 100
print(units_sum_by_royaly)
```

### вычисляю стоимость одного юнита по площадкам

```python
unit_price = units_sum_by_royaly / units_sum_by_outletname
print(type(unit_price))
unit_price = unit_price.sort_values(ascending = False)

print(unit_price)

```

### построить график

```python
unit_price.plot.barh();
```
